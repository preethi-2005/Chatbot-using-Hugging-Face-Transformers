## **Step 2**: Importing Libraries



```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

## **Step 3**: Loading Pre-trained Model (gpt2)




```python
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token

print("GPT-2 Model loaded successfully!\n")
```

## **Step 4:** Chatbot Implementation

  

```python
import datetime

conversation_history = ""

print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

while True:
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Conversation ended. Goodbye!")
        break

    # RULE-BASED HANDLING
    # Time query
    if "time" in user_input.lower():
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        print(f"Chatbot: The current time is {current_time}")
        continue

    # Date query
    if "date" in user_input.lower():
        current_date = datetime.datetime.now().strftime("%Y-%m-%d")
        print(f"Chatbot: Today's date is {current_date}")
        continue

    # Greeting
    if user_input.lower() in ["hi", "hello", "hey"]:
        print("Chatbot: Hello! How can I assist you today?")
        continue

    # Conversation memory
    conversation_history += f"User: {user_input}\n"

    prompt = (
        "You are a helpful AI assistant.\n"
        "Answer ONLY what is asked.\n"
        "Be short, clear, and relevant.\n"
        "Do NOT generate extra questions or unrelated content.\n"
        "If unsure, say you are not certain.\n\n"
        f"{conversation_history}"
        "Assistant:"
    )

    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Output generation
    output = model.generate(
        input_ids,
        max_length=input_ids.shape[1] + 60,
        do_sample=False,
        repetition_penalty=1.2,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(
        output[:, input_ids.shape[-1]:][0],
        skip_special_tokens=True
    ).strip()

    response = response.split("User:")[0].strip()

    if not response:
        response = "I'm not completely sure. Could you rephrase your question?"


    conversation_history += f"Assistant: {response}\n"

    print("Chatbot:", response)
```
